Virat Kohli Stats

```python
virat = data[data['batsman'] == 'V Kohli']
virat.shape
```

```python
virat['bowling_team'] = virat['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
virat['bowling_team'] = virat['bowling_team'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
virat['bowling_team'].unique()
```

```python
virat.sample(5)
```

```python
total_runs = virat['batsman_runs'].sum()
print("Total_runs : ",total_runs)
balls_faced = virat['batsman_runs'].count()
print('Balls faced : ',balls_faced)
total_matches = len(virat['match_id'].unique())
dismisal_count  = virat['player_dismissed'].count()
print("Dismissed number : ",dismisal_count)
print("Total matches : ",total_matches)
print("Average : ",total_runs/dismisal_count)
```

```python
matches = virat.drop_duplicates(subset=['match_id'])
matches.sample(5)
```

```python
matches['bowling_team'] = matches['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
matches['bowling_team'] = matches['bowling_team'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
matches['bowling_team'].unique()
```

```python
matches_with_each_team = matches['bowling_team'].value_counts()
matches_with_each_team
```

```python
runs_per_team = virat.groupby(['bowling_team']).sum()
runs_per_team = runs_per_team['batsman_runs']
runs_per_team
```

```python
avg_per_team = runs_per_team/matches_with_each_team
avg_per_team
```
